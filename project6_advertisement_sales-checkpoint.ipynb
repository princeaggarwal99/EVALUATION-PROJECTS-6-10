```python
## treating outliers
from scipy.stats import zscore
```

```python
z=np.abs(zscore(ad_data))
print(np.where(z>3))  ## 3 is the threshold value
```

```python
ad_new=ad_data[(z<3).all(axis=1)]
ad_new.head()
```

```python
ad_new.shape
## 1 row reduced due to removing outlier
```

```python
## seperating target variable
x=ad_new.iloc[:,0:-1]
y=ad_new.iloc[:,-1] ## target variable is y
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
```

```python
x=ss.fit_transform(x)
```

```python
x# tv,radio,newspaper (features)
```

```python
y.head() #target variable
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
## function for finding the best random state for the LINEAR REGRESSION
max_r_score=0
for r_state in range(42,101):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r_state,test_size=0.22)
    lr=LinearRegression()
    lr.fit(x_train,y_train)
    ylr=lr.predict(x_test)
    r2_scr=r2_score(y_test,ylr)
    print('r2_score corr to random state',r_state,'is:',r2_scr)
    if r2_scr>max_r_score:
        max_r_score=r2_scr
        final_r_state=r_state
print()
print()
print("max r2 score corr to ",final_r_state,'is',max_r_score)       
        
```

```python
## printing the shape of train and test set
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
# best random state=90  for model
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=90,test_size=0.22)
lr=LinearRegression()
lr.fit(x_train,y_train)
ylr=lr.predict(x_test)
r2_scr=r2_score(y_test,ylr)
print(mean_squared_error(ylr, y_test))
print(mean_absolute_error(ylr, y_test))
print(cross_val_score(lr,x,y,cv=10,scoring='r2').mean())
print(np.sqrt(mean_squared_error(ylr, y_test)))
```
